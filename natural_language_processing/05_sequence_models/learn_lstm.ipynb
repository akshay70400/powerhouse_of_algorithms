Long Short Term Memory

For every time stamp, it has two states:
    Memory Cell state - Ct
    Hidden state - ht
    
The reason why it is able to preserve the context of the words for so long is due to Gates:
    1. Forget gate - responsible for retaining or forgetting certain information
    2. Input gate - responsible to add new info
    3. Output gate - 
Gate allows the certain information to pass through and restricts the other information.
