# MERGE

```python
pd.merge()
```

```python
data1 = {'key': ['K0', 'K1', 'K2', 'K3'],
         'Name':['Jai', 'Princi', 'Gaurav', 'Anuj'], 
        'Age':[27, 24, 22, 32],} 

data2 = {'key': ['K0', 'K1', 'K2', 'K3'],
         'Address':['Nagpur', 'Kanpur', 'Allahabad', 'Kannuaj'], 
        'Qualification':['Btech', 'B.A', 'Bcom', 'B.hons']} 
```

```python
df_a = pd.DataFrame(data1)
df_b= pd.DataFrame(data2)
```

```python
df_a
```

```python
df_b
```

```python
res1=pd.merge(df_a,df_b, on="key", how="inner")

res1
```

```python
data1 = {'key': ['K0', 'K1', 'K2', 'K3'],
         'key1': ['K0', 'K1', 'K0', 'K1'],
         'Name':['Jai', 'Princi', 'Gaurav', 'Anuj'], 
        'Age':[27, 24, 22, 32],} 

data2 = {'key': ['K0', 'K1', 'K2', 'K3'],
         'key1': ['K0', 'K0', 'K0', 'K0'],
         'Address':['Nagpur', 'Kanpur', 'Allahabad', 'Kannuaj'], 
        'Qualification':['Btech', 'B.A', 'Bcom', 'B.hons']} 
```

```python
df_a = pd.DataFrame(data1)
df_b= pd.DataFrame(data2)
```

```python
df_a
```

```python
df_b
```

```python
res1=pd.merge(df_a,df_b, on=["key","key1"], how="inner")

res1
```

```python
res1=pd.merge(df_a,df_b, on=["key","key1"], how="outer")

res1
```

```python
res1=pd.merge(df_a,df_b, on=["key","key1"], how="left")

res1
```

```python
res1=pd.merge(df_a,df_b, on=["key","key1"], how="right")

res1
```

# JOIN

```python
data1 = {'Name':['Jai', 'Princi', 'Gaurav', 'Anuj'], 
        'Age':[27, 24, 22, 32]} 

data2 = {'Address':['Allahabad', 'Kannuaj', 'Pune', 'Kannuaj'], 
        'Qualification':['MCA', 'Phd', 'Bcom', 'B.hons']}
```

```python
j1= pd.DataFrame(data1,index=['K0', 'K1', 'K2', 'K3'])

j2 = pd.DataFrame(data2, index=['K0', 'K2', 'K3', 'K4'])
```

```python
j1
```

```python
j2
```

```python
res=j1.join(j2, how="left")
res
```

```python
res=j1.join(j2, how="right")
res
```

```python
res=j1.join(j2, how="outer")
res
```

```python
res=j1.join(j2, how="inner")
res
```
