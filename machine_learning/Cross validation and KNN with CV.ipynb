```python
# Hold out

from sklearn.linear_model import LogisticRegression as LG
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test=train_test_split(x,y,test_size=0.30,random_state=4)

model=LG(max_iter=300)
model.fit(x_train,y_train)
result=model.score(x_test,y_test)
print(result)
```

```python
# K folf cross validation
```

```python
import numpy as np
from sklearn.model_selection import KFold
model=LG(max_iter=300,random_state=100,)
kfold_validation=KFold(10)

from sklearn.model_selection import cross_val_score

result=cross_val_score(model,x,y,cv=kfold_validation)
print("All accuracy",result)

print("mean",np.mean(result))
```

# leave one out

```python
import numpy as np
from sklearn.model_selection import LeaveOneOut
model=LG(max_iter=300,random_state=100,)
Leave_validation=LeaveOneOut()

from sklearn.model_selection import cross_val_score

result=cross_val_score(model,x,y,cv=Leave_validation)
print("All accuracy",result)

print("mean",np.mean(result))
```

```python
import numpy as np
from sklearn.model_selection import ShuffleSplit
model=LG(max_iter=300,random_state=100,)
Sl=ShuffleSplit()

from sklearn.model_selection import cross_val_score

result=cross_val_score(model,x,y,cv=Sl)
print("All accuracy",result)

print("mean",np.mean(result))
```

```python
import numpy as np
from sklearn.model_selection import StratifiedKFold
model=LG(max_iter=300,random_state=100,)
Skfold=ShuffleSplit()

from sklearn.model_selection import cross_val_score

result=cross_val_score(model,x,y,cv=Skfold)
print("All accuracy",result)

print("mean",np.mean(result))
```
